```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab import files

print("Silakan unggah (upload) gambar Anda:")
uploaded = files.upload()

filename = next(iter(uploaded))
print(f"\nMemproses gambar: {filename}")

img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
_, img_biner = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

kernel = np.ones((5,5), np.uint8)

erosi = cv2.erode(img_biner, kernel, iterations=1)
opening = cv2.morphologyEx(img_biner, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img_biner, cv2.MORPH_CLOSE, kernel)

titles = ['Gambar Biner (Input Asli)', '1. Hasil Erosi', '2. Hasil Opening', '3. Hasil Closing']
images = [img_biner, erosi, opening, closing]

plt.figure(figsize=(15, 10))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i], fontsize=14)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
